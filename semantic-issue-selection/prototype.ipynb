```python
from sentence_transformers import SentenceTransformer, util

query = "The driver was using inappropriate language during the ride, making me uncomfortable. Please address this."


def get_most_similar(query, reference_dict):
    # Load the model
    model = SentenceTransformer("sentence-transformers/multi-qa-mpnet-base-cos-v1")

    references = list(reference_dict.values())
    issue_types = list(reference_dict.keys())

    # Encode query and documents
    query_emb = model.encode(query)
    doc_emb = model.encode(references)

    # Compute dot score between query and all document embeddings
    scores = util.dot_score(query_emb, doc_emb)[0].cpu().tolist()

    # Combine docs & scores
    issue_types_score_pairs = list(zip(issue_types, scores))

    # Sort by decreasing score
    issue_types_score_pairs = sorted(issue_types_score_pairs, key=lambda x: x[1], reverse=True)

    # Output passages & scores
    return {issue_type: score for issue_type, score in issue_types_score_pairs}
```

```python
get_most_similar(query, REFERENCE_DICT)
```

```python
from collections import defaultdict


def predict_issue_type():
    eval_dict = defaultdict(list)
    for issue_type, samples in ISSUES_DICT.items():
        for sample in samples:
            eval_dict["issue_type"].append(issue_type)
            eval_dict["sample"].append(sample)
            semantically_similar = get_most_similar(sample, REFERENCE_DICT)
            eval_dict["full_results"].append(semantically_similar)
            eval_dict["top_match"].append(list(semantically_similar.keys())[0])
            eval_dict["top_score"].append(list(semantically_similar.values())[0])
            eval_dict["top_3_match"].append(list(semantically_similar.keys())[0:3])

    df = pd.DataFrame(eval_dict)
```

```python
df["top_accuracy"] = (df["issue_type"] == df["top_match"]).astype(int)
df["top_3_accuracy"] = df.apply(lambda x: x["issue_type"] in x["top_3_match"], axis=1).astype(int)
```
